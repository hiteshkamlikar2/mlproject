```python
import pandas as pd

#Read the data and create a copy

Data= pd.read_csv("C:\\Users\\2004674\\Downloads\\04+-+decisiontreeAdultIncome.csv")
# LoanData.head()
display(Data)
LoanPrep = Data.copy()
#check for null values
Data.isnull().sum(axis=0)
```

```python
#create get dummy variables
dat_prep = pd.get_dummies(Data,drop_first=True)
#create the x and y variable
x = dat_prep.iloc[:,:-1]
y=dat_prep.iloc[:,-1]
```

```python
#train and test the data set

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = \
train_test_split(x,y,test_size=0.3,random_state=1234,stratify=y)
```

```python
from sklearn.ensemble import RandomForestClassifier
#train Model 
rfc = RandomForestClassifier(random_state=1234)
rfc.fit(x_train,y_train)
y_predict = rfc.predict(x_test)
```

```python
# confusion Matrix
from sklearn.metrics import confusion_matrix

cm2 = confusion_matrix(y_test,y_predict)
score2=rfc.score(x_test,y_test)
print(score2)
```
