```python
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
diabetes = pd.read_csv("C:/Users/2004674/Downloads/diabetes.csv")

diabetes.isnull().sum(axis=0)
```

```python
x = diabetes.iloc[:,0:-1]
y= diabetes.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = \
train_test_split(x,y,test_size=0.2,random_state=1234, stratify=y)
```

```python
# define the sequential model

model = Sequential()

model.add(Dense(24,
                input_shape=(8,),
               activation='relu',
                kernel_initializer='RandomNormal'))

# Add is method to add layer called Dense, 24 is number of unit you want to
# to add  and input_shape = 8 means in x we have 8 features so we added 8

model.add(Dense(12,
                activation='relu',
                kernel_initializer='RandomNormal'))

# output layer as we are dealing with binary layer we need only 1 nueron
model.add(Dense(1,
                activation='sigmoid',
                kernel_initializer='RandomNormal'))
```

```python
# fit and train/compile the model
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(x_train,y_train,epochs=160,batch_size=10)
```

```python
# Evalute the model

accuracy_test = model.evaluate(x_test,y_test)

#confusion matrix
y_pred_prob = model.predict(x_test)

predictions = (y_pred_prob > 0.5).astype('int32')
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,predictions)
```
