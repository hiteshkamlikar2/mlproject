```python
# Implemented Recursive Feature Eliminatin.
# predict product purchase for tha bank telemarketing 
import pandas as pd

f = pd.read_csv("C:\\Users\\2004674\\Downloads\\bank.csv")

# drop the feature duration beacause it has no importance

f = f.drop("duration",axis=1)
```

```python
# split the columns into dependent (y) and independent (x)

x = f.iloc[:,:-1]
y = f.iloc[:,-1]
```

The get_dummies function works as follows:

It takes a data frame, series, or list.
Then, it converts each unique element present in the object to a column heading.
The function iterates over the object that is passed and checks if the element at the particular index matches the column heading.
If it does, it encodes it as a 1.

Otherwise, it assigns it a 0.

```python
# create the dummy variables
# The get_dummies function is used to convert categorical variables into dummy or indicator variables.
x = pd.get_dummies(x,drop_first=True)
y - pd.get_dummies(y,drop_first=True)
```

```python
# split the dataset into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = \
train_test_split (x,y,test_size=0.3 ,random_state=1234,stratify=y)
```

```python
# import random forest classifier 
from sklearn.ensemble import RandomForestClassifier

# Default random forest object
rfc1 = RandomForestClassifier(random_state=1234)
rfc1.fit(x_train,y_train)
Y_predict = rfc1.predict(x_test)
```

```python
# score and Evaluate the model

from sklearn.metrics import confusion_matrix
cm1 = confusion_matrix(y_test,Y_predict)
score1 = rfc1.score(x_test,y_test)
```

estimatorEstimator instance
A supervised learning estimator with a fit method that provides information about feature importance (e.g. coef_, feature_importances_).

n_features_to_selectint or float, default=None
The number of features to select. If None, half of the features are selected. If integer, the parameter is the absolute number of features to select. If float between 0 and 1, it is the fraction of features to select.

stepint or float, default=1
If greater than or equal to 1, then step corresponds to the (integer) number of features to remove at each iteration. If within (0.0, 1.0), then step corresponds to the percentage (rounded down) of features to remove at each iteration.

```python
# import RFE

from sklearn.feature_selection import RFE

# get an estimator object 
rfc2 = RandomForestClassifier(random_state=1234)

rfe = RFE(estimator=rfc2, n_features_to_select=30, step = 1)

rfe.fit(x,y)
```

```python
# create new input data set and transorm
# here we get the rfe data into x train and test as 30 columns
# before it was 52 columns and also our column name is remove
x_train_rfe = rfe.transform(x_train)
x_test_rfe = rfe.transform(x_test)
```

```python
# Default random forest object
# fit the random forest classifier to the new train and test with 80 features
rfc2.fit(x_train_rfe, y_train)

# test the model with new test dataset
Y_predict = rfc2.predict(x_test_rfe)

# score and Evaluate the model

from sklearn.metrics import confusion_matrix
cm_rfe = confusion_matrix(y_test,Y_predict)
score_rfe = rfc2.score(x_test_rfe,y_test)
```

```python
# get all the columns names
columns = list(x.columns)

# ranking 1 for selected features
ranking = rfe.ranking_

# get the features importance score
# here rfc1 beacause rfc1 has all feature whereas rfc2 has only important feartures
feature_importance = rfc1.feature_importances_

# create the dataframe of the selected,ranking and features importance

rfe_selected = pd.DataFrame()
rfe_selected =pd.concat([pd.DataFrame(columns),pd.DataFrame(ranking),pd.DataFrame(feature_importance)],axis = 1)
# as we have array and list so concate will give us error so we convetred all into dataframes

rfe_selected.columns = ["feature names","Ranking","Feature Importance"]
```
