```python
# impleneted Ridge regression  with different values of lambda/alpha

import pandas as pd 
dataset = pd.read_csv("C:\\Users\\2004674\\Downloads\\ridge.csv")

df = dataset.copy()
```

```python
# split the data into x (independent ) and y (predicted)
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
# ridge regression
from sklearn.linear_model import Ridge
l =1 # l is alpha values
ridge = Ridge(alpha=l)
ridge.fit(x , y)

# get the coefficient
ridge_coeeff = ridge.coef_

# get the interecept
rigde_intecept = ridge.intercept_
```

```python
# ols alpha as 0 y = 2x+0
# ridge as alpha =1 , y = 1.667x + 0.833
#0.833 is intercept and 1.66 is coeeficient
```

```python
# plot the Ridge
# x_plt = [0,1,2,3,4]
# plot the Ridge
x_plt = [0,1,2,3,4]

lasso_l = [0,1,10,100]
```

```python
import matplotlib.pyplot as plt

for i , l in enumerate(lasso_l):
    ridge = Ridge(alpha=l)
    ridge.fit(x , y)
    # get the coefficient
    ridge_coeeff = ridge.coef_

# get the interecept
    ridge_intecept = ridge.intercept_
    y_plt = ridge.predict(pd.DataFrame(x_plt))

    plt.figure(1)
    plt.plot(x_plt, y_plt)
    plt.ylim(ymin=0,ymax=9) 
    plt.xlim(xmin=0,xmax=6) 
    plt.text(x_plt[-1],y_plt[-1],
             'y = '
              + str('%.2f'%ridge_intecept) 
              + '* x '
              + str('%.2f'%ridge_intecept)
              
    )
```
