```python
# impleneted Ridge regression  with different values of lambda/alpha

import pandas as pd 
dataset = pd.read_csv("C:\\Users\\2004674\\Downloads\\ridge.csv")

df = dataset.copy()
```

```python
# split the data into x (independent ) and y (predicted)
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
# ridge regression
from sklearn.linear_model import Lasso
l =1 # l is alpha values
lasso = Lasso(alpha=l)
lasso.fit(x , y)

```

```python
# plot the Ridge
x_plt = [0,1,2,3,4]

lasso_l = [0,0.5,1,2,4]
```

```python
import matplotlib.pyplot as plt

for i , l in enumerate(lasso_l):
    lasso = Lasso(alpha=l)
    lasso.fit(x , y)
    # get the coefficient
    lasso_coeeff = lasso.coef_

# get the interecept
    lasso_intecept = lasso.intercept_
    y_plt = lasso.predict(pd.DataFrame(x_plt))

    plt.figure(1)
    plt.plot(x_plt, y_plt)
    plt.ylim(ymin=0,ymax=9) 
    plt.xlim(xmin=0,xmax=6) 
    plt.text(x_plt[-1],y_plt[-1],
             'y = '
              + str('%.2f'%lasso_coeeff) 
              + '* x '
              + str('%.2f'%lasso_intecept)
              
    )
```
